# Importing the dependencies

```python
print("Importing dependencies...")
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import os
print("Dependencies loaded successfully!")
```

## Data Collection and Visualization

```python
filename = "diabetes.csv"
dirname = "datasets"
path = os.path.join(dirname, filename)

df = pd.read_csv(path)
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.groupby("Outcome").mean()
```

0 -> diabetic
1 -> non diabetic

```python
Y = df.Outcome
```

```python
print(Y.shape)
```

```python
X = df.drop(columns="Outcome")
print(X.shape)
print(X.head())
```

## Create and train Models

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.fit_transform(X) # Standardize the X
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=1)
```

```python
# train model
model = svm.SVC(kernel="linear")
model.fit(X_train, Y_train)

# validation
predictions = model.predict(X_test)
print(X_scaled)
score = accuracy_score(predictions, Y_test)
score = round(score*100)

print(f"score: {score}%")
```

```python
# test the model
data_test = [1, 85, 66,	29,	0, 26.6, 0.351,	31]
d = np.asarray(data_test)
d = d.reshape(1, -1)
print(d)

# standardize data
d_std = scaler.transform(d)
```

```python
prediction = model.predict(d_std)
response = "Diabetic" if prediction else "Non diabetic"
print(response)
```
